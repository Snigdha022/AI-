```python
import re

def read_input_file(file_name):
    facts = []
    rules = []
    with open(file_name, 'r') as file:
        for line in file:
            line = line.strip()
            if line.find(' => ') != -1:
                rule = re.split(' => ', line)
                rule_premise = re.split(' \^ ', rule[0])
                rules.append((rule_premise, rule[1]))
            else:
                facts.append(line)
    print('rules:')
    print(rules)
    return facts, rules

def forward_chaining(facts, rules, fact_to_find):
    
    new_facts = set(facts)
    while True:
        for premise, conclusion in rules:
            
            print('premise:')
            print(premise)
            print('conclusion:')
            print(conclusion)
            if set(premise).issubset(new_facts):
                new_facts.add(conclusion)
                print('new_facts:')
                print(new_facts)
        if fact_to_find in new_facts:
            return True
        
        if set(new_facts) == set(facts):
            return False
        facts = set(new_facts)

facts, rules = read_input_file("/content/input.txt")

fact_to_find = facts.pop()
print('facts:')
print(facts)
result = forward_chaining(facts, rules, fact_to_find)
if result:
    print(f"{fact_to_find} Found")
else:
    print(f"{fact_to_find} Not Found")
```
